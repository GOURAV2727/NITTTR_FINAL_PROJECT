# Importing Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("Parkinsson disease.csv")
```

## RENAMING COLUMNS

```python
df.rename(columns=({'MDVP:Fo(Hz)':'avg_fre', 'MDVP:Fhi(Hz)':'max_fre', 'MDVP:Flo(Hz)':'min_fre', 'MDVP:Jitter(%)':'var_fre1',
       'MDVP:Jitter(Abs)':'var_fre2', 'MDVP:RAP':'var_fre3', 'MDVP:PPQ':'var_fre4', 'Jitter:DDP':'var_fre5',
       'MDVP:Shimmer':'var_amp1', 'MDVP:Shimmer(dB)':'var_amp2', 'Shimmer:APQ3':'var_amp3', 'Shimmer:APQ5':'var_amp4',
       'MDVP:APQ':'var_amp5', 'Shimmer:DDA':'var_amp6'}),inplace=True)
df
```

## PREPROCESSING 

```python
df.describe()
```

```python
df.isnull().any(axis=0)
```

```python
df=df.drop(columns=['name'],axis=1)
print(df)
```

## FINDING THE OUTLIERS WITH HELP OF GRAPH


```python
x1=df.loc[:,df.columns!='status']
x1.hist(figsize=(25,20))
plt.show()
```

## REMOVING OUTLIERS

```python

df=df[df.max_fre<=300]
df=df[df.var_fre1<=0.02]
df=df[df.var_fre2<=0.0001]
df=df[df.var_fre3<=0.01]
df=df[df.var_fre4<=0.01]
df=df[df.var_fre5<=0.02]
df=df[df.var_amp1<=0.10]
df=df[df.var_amp2<=1.0]
df=df[df.var_amp3<=0.04]
df=df[df.var_amp4<=0.050]
df=df[df.var_amp5<=0.075]
df=df[df.var_amp6<=0.125]
df=df[df.NHR<=0.15]

```

## HEATMAP

```python
cor=x1.corr()
plt.figure(figsize=(20,20))
sns.heatmap(cor,annot=True)
plt.show()
```

## DEFINING DEPENDENT AND INDEPENDENT VARIABLE

```python
x=df.loc[:,df.columns!='status'].values
y=df.loc[:,'status'].values
```

### Normalizing the data 

```python
# normalizing the data between (0-1) to process the data easily
from sklearn.preprocessing import StandardScaler
sk=StandardScaler()
x1=sk.fit_transform(x)
```

### Spliting the data into Train and Test

```python
import sklearn
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest=train_test_split(x,y,random_state=0,test_size=.30)
```

### DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
classifier=DecisionTreeClassifier()
classifier.fit(xtrain,ytrain)
predict=classifier.predict(xtest)
accuracy_score(ytest, predict)*100
```

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(xtrain,ytrain)
predict=lg.predict(xtest)
accuracy_score(predict,ytest)*100
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
rd=RandomForestClassifier()
rd.fit(xtrain,ytrain)
predict=rd.predict(xtest)
accuracy_score(predict,ytest)*100
```

### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(xtrain,ytrain)
y_pred = knn.predict(xtest)
accuracy_score(ytest,y_pred)*100
```

### AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
a=AdaBoostClassifier()
a.fit(xtrain,ytrain)
predict=a.predict(xtest)
accuracy_score(predict,ytest)*100
```

### GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier
b=GradientBoostingClassifier()
b.fit(xtrain,ytrain)
predict=a.predict(xtest)
accuracy_score(predict,ytest)*100
```

### Confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,predict)
plt.figure(figsize=(8,6))
fg=sns.heatmap(cm,annot=True)
figure=fg.get_figure()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Output Confusion Matrix")
```
